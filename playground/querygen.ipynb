```python
from openai import OpenAI

client = OpenAI(base_url="http://localhost:8000/v1", api_key="dummy")
prompts = [
    "Solve this math problem step by step: What is the integral of x^2 from 0 to 1? Please reason step by step and put your final answer within \\boxed{}.",
    "Write a short story about a robot learning to paint.",
    "Is 561 a prime number? Reason step by step and box the final answer."
]

def send_request(prompt):
    return client.completions.create(
        model=model,
        prompt=prompt,
        temperature=0.6,
        max_tokens=2048
    ).choices[0].text

# Process prompts sequentially
results = []
for prompt in prompts:
    result = send_request(prompt)
    results.append(result)

# Print results
for prompt, result in zip(prompts, results):
    print(f"Prompt: {prompt}\nResponse: {result}\n{'-'*50}")
```
